## üìò 1.1 ‚Äì Funci√≥n de transferencia del sistema masa-resorte-amortiguador

### üîß Modelo f√≠sico

El sistema masa-resorte-amortiguador est√° compuesto por:

- Una **masa** que se mueve en una dimensi√≥n.
- Un **resorte** que ejerce una fuerza proporcional al desplazamiento.
- Un **amortiguador viscoso** que ejerce una fuerza proporcional a la velocidad.

Aplicando la **segunda ley de Newton** (fuerza neta igual a masa por aceleraci√≥n), se obtiene la ecuaci√≥n diferencial:

$$
m \frac{d^2x(t)}{dt^2} + b \frac{dx(t)}{dt} + kx(t) = F(t)
$$

Donde:
- $m$ es la masa del cuerpo (kg)
- $b$ es el coeficiente de amortiguamiento (N¬∑s/m)
- $k$ es la constante del resorte (N/m)
- $x(t)$ es el desplazamiento (m)
- $F(t)$ es la fuerza externa aplicada (N)


### Transformada de Laplace

Aplicamos la transformada de Laplace a ambos lados de la ecuaci√≥n, **suponiendo condiciones iniciales iguales a cero**:

- $\mathcal{L}\left\{\frac{d^2x(t)}{dt^2}\right\} = s^2 X(s)$
- $\mathcal{L}\left\{\frac{dx(t)}{dt}\right\} = s X(s)$
- $\mathcal{L}\left\{x(t)\right\} = X(s)$
- $\mathcal{L}\left\{F(t)\right\} = F(s)$

Entonces, la ecuaci√≥n diferencial se transforma en:

$$
m s^2 X(s) + b s X(s) + k X(s) = F(s)
$$



### Despejar la funci√≥n de transferencia

Factorizamos $X(s)$ en el lado izquierdo:

$$
X(s) \left( m s^2 + b s + k \right) = F(s)
$$

Dividimos ambos lados entre el polinomio:

$$
\frac{X(s)}{F(s)} = \frac{1}{m s^2 + b s + k}
$$


### Funci√≥n de transferencia en lazo abierto

$$
H(s) = \frac{X(s)}{F(s)} = \frac{1}{m s^2 + b s + k}
$$

###  Analog√≠as entre mec√°nica y electricidad

| Sistema Mec√°nico | Sistema El√©ctrico (RLC serie) |

### Ecuaci√≥n del circuito RLC en serie

Supongamos un circuito RLC en serie, con entrada $V(t)$ y salida la **carga del capacitor** $q(t)$, y con corriente $i(t) = \frac{dq(t)}{dt}$. Aplicamos la ley de voltajes de Kirchhoff:

$$
V(t) = L \frac{d^2q(t)}{dt^2} + R \frac{dq(t)}{dt} + \frac{1}{C} q(t)
$$


### Aplicar transformada de Laplace

Nuevamente, aplicamos transformada de Laplace asumiendo condiciones iniciales cero:

- $\mathcal{L}\left\{\frac{d^2q(t)}{dt^2}\right\} = s^2 Q(s)$
- $\mathcal{L}\left\{\frac{dq(t)}{dt}\right\} = s Q(s)$
- $\mathcal{L}\left\{q(t)\right\} = Q(s)$

Entonces:

$$
V(s) = L s^2 Q(s) + R s Q(s) + \frac{1}{C} Q(s)
$$



### Despejar la funci√≥n de transferencia

Factorizamos $Q(s)$:

$$
Q(s) \left( L s^2 + R s + \frac{1}{C} \right) = V(s)
$$

Dividimos ambos lados:

$$
\frac{Q(s)}{V(s)} = \frac{1}{L s^2 + R s + \frac{1}{C}}
$$



###  Funci√≥n de transferencia equivalente (circuito el√©ctrico)

$$
H(s) = \frac{Q(s)}{V(s)} = \frac{1}{L s^2 + R s + \frac{1}{C}}
$$


# Punto 2 ‚Äì Modulaci√≥n y demodulaci√≥n por amplitud en banda lateral √∫nica (SSB-AM)

## Introducci√≥n

La modulaci√≥n en banda lateral √∫nica (SSB-AM) es una t√©cnica de modulaci√≥n en amplitud que elimina una de las bandas laterales del espectro de frecuencia, y en algunos casos tambi√©n la portadora, para reducir el ancho de banda y mejorar la eficiencia energ√©tica del sistema de transmisi√≥n.

Existen dos tipos principales de modulaci√≥n SSB:
- **SSB con portadora suprimida**: se transmite solo una banda lateral (superior o inferior).
- **SSB con portadora reducida o presente**: se transmite la portadora con menor amplitud junto con una sola banda lateral.



## Se√±al modulada en el dominio del tiempo

La se√±al modulada SSB-AM puede representarse en el dominio del tiempo de dos formas equivalentes: una basada en la descomposici√≥n de Hilbert y otra basada en la expresi√≥n trigonom√©trica.

### Representaci√≥n con transformada de Hilbert

Sea $m(t)$ la se√±al mensaje (real), y $\hat{m}(t)$ su transformada de Hilbert. Entonces, la se√±al SSB puede expresarse como:

- **Banda lateral superior (USB):**
  $$
  s_{\text{USB}}(t) = m(t) \cos(2\pi f_c t) - \hat{m}(t) \sin(2\pi f_c t)
  $$

- **Banda lateral inferior (LSB):**
  $$
  s_{\text{LSB}}(t) = m(t) \cos(2\pi f_c t) + \hat{m}(t) \sin(2\pi f_c t)
  $$

Donde:
- $f_c$ es la frecuencia de la portadora.
- $\hat{m}(t)$ es la transformada de Hilbert de $m(t)$.



## Se√±al modulada en el dominio de la frecuencia

Sea $M(f)$ la transformada de Fourier de la se√±al mensaje $m(t)$. La se√±al DSB (doble banda lateral) tiene el espectro:

$$
S_{\text{DSB}}(f) = \frac{1}{2} \left[ M(f - f_c) + M(f + f_c) \right]
$$

En SSB, se elimina una de las bandas laterales:

- **SSB-USB (banda superior):**
  $$
  S_{\text{USB}}(f) = M(f - f_c), \quad f \geq f_c
  $$

- **SSB-LSB (banda inferior):**
  $$
  S_{\text{LSB}}(f) = M(f + f_c), \quad f \leq -f_c
  $$

Por lo tanto, el espectro de la se√±al SSB ocupa la mitad del ancho de banda de una se√±al DSB y transmite la misma cantidad de informaci√≥n.



## Proceso de demodulaci√≥n

Para recuperar $m(t)$ a partir de $s_{\text{SSB}}(t)$, se puede usar la t√©cnica de **demodulaci√≥n coherente**, que requiere una portadora sincronizada en fase y frecuencia con la portadora original. Por ejemplo, en el caso USB:

1. Multiplicaci√≥n por $\cos(2\pi f_c t)$:
   $$
   s_{\text{USB}}(t) \cdot \cos(2\pi f_c t)
   $$

2. Paso por un filtro pasa bajos ideal con ancho de banda suficiente para cubrir el contenido de $m(t)$.

Tambi√©n es posible aplicar el operador de demodulaci√≥n usando la **transformada de Hilbert inversa**, si se dispone del canal en fase y en cuadratura.



## Aplicaci√≥n de filtros digitales IIR

Para implementar el sistema SSB de forma digital, se requiere:

- Un **filtro de Hilbert** (para obtener $\hat{m}(t)$).
- Filtros **pasa banda** o **pasa bajos** digitales para separar bandas.
- Durante la demodulaci√≥n, un filtro **pasa bajos** IIR se usa para recuperar $m(t)$ desde la se√±al modulada.

Los filtros **IIR (Infinite Impulse Response)** son preferidos en muchas implementaciones por su eficiencia computacional y buena selectividad con menor orden, aunque su dise√±o requiere cuidado para evitar inestabilidades.



## Visualizaci√≥n y an√°lisis en frecuencia

- **Plano de polos y ceros**, que permite evaluar la estabilidad y la respuesta en frecuencia del filtro.

Estos elementos son claves para comprender el efecto del filtrado sobre la se√±al y garantizar que la modulaci√≥n y demodulaci√≥n sean efectivas.
